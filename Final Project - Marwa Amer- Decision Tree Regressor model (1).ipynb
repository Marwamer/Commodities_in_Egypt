```python
data['year'] = data['date'].dt.year
cat_grouped = data.groupby(['category', 'year'])
cat_result = cat_grouped.mean()
cat_result = cat_result.reset_index()
cat_result
```

```python
data['year'] = data['date'].dt.year
com_grouped = data.groupby(['commodity', 'year', 'unit'])
com_result = com_grouped.mean()
com_result = com_result.reset_index()
com_result
```

```python

sns.set_style('darkgrid')

sns.lineplot(x='year', y='price', hue='category', data=cat_result)

plt.title('Price Change over Time by Category')
plt.xlabel('Year')
plt.ylabel('Price')

plt.show()
```

```python
import matplotlib.pyplot as plt

avg_price = com_result.groupby('commodity')['price'].mean()

plt.bar(avg_price.index, avg_price.values)

plt.title('Average Price by Commodity')
plt.xlabel('Commodity')
plt.ylabel('Price')

plt.xticks(rotation=90)

plt.show()
```

```python
pivot_df = com_result.pivot('commodity', 'year', 'price')

sns.heatmap(pivot_df, cmap='coolwarm')
plt.title('Price by Commodity and Year')
plt.xlabel('Year')
plt.ylabel('Commodity')
plt.show()
```
